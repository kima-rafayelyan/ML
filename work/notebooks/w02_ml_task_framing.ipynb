## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*


Scoring (ranking). My goal is to assign each page a priority score based on its performance metrics. Pages with higher scores should be reviewed first because they show stronger signs that content updates may be valuable. This approach creates a ranked list of pages instead of placing them into fixed categories, making it easier for content teams to decide where to focus their efforts first.

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")
print(df.shape)
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

I would predict each page's refresh priority score and suggested action (refresh, expansion, protection, pruning, or monitoring). The label comes from a defined rule rather than an observed outcome. For example, pages can be scored using factors such as declining traffic, low CTR, outdated content, or weak engagement, and the score determines the recommended action.
